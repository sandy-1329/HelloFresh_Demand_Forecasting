```python
# Demand Forecasting for HelloFresh-style Meals
# Real-world dataset from Kaggle (cleaned)

import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from pmdarima import auto_arima
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Load cleaned dataset
df = pd.read_csv('/content/meal_demand_cleaned.csv')

df['week'] = pd.to_datetime(df['week'])

# Focus on one popular meal for time series modeling
meal_id = df['meal_id'].value_counts().idxmax()
meal_df = df[df['meal_id'] == meal_id][['week', 'num_orders']].sort_values('week')
meal_df.rename(columns={'week': 'ds', 'num_orders': 'y'}, inplace=True)

# Plot demand over time
plt.figure(figsize=(10, 5))
plt.plot(meal_df['ds'], meal_df['y'], marker='o')
plt.title(f"Weekly Demand for Meal ID {meal_id}")
plt.xlabel("Week")
plt.ylabel("Number of Orders")
plt.grid(True)
plt.show()

# -----------------------------
# Prophet Model
# -----------------------------
print("Running Facebook Prophet...")

prophet = Prophet()
prophet.fit(meal_df)

# Forecast next 12 weeks
future = prophet.make_future_dataframe(periods=12, freq='W')
forecast = prophet.predict(future)

# Plot forecast
prophet.plot(forecast)
plt.title(f"Forecast for Meal ID {meal_id} using Prophet")
plt.xlabel("Date")
plt.ylabel("Orders")
plt.show()

# Evaluation on last 12 known weeks
actual = meal_df.set_index('ds').y[-12:].values
# Note: Prophet forecasts include the historical data points used for training.
# To get the forecast for the last 12 *actual* weeks in the training data,
# we need to select the corresponding rows from the forecast dataframe.
# The forecast dataframe contains predictions for all dates in the 'future' dataframe.
# The 'future' dataframe starts with the dates from meal_df['ds'] and appends the future dates.
# Therefore, the predictions for the last 12 actual weeks will be in the rows
# corresponding to the last 12 dates from the original meal_df['ds'].
# The total number of data points in meal_df is len(meal_df).
# The forecast dataframe has len(meal_df) + periods rows.
# The predictions for the last 12 actual weeks are at indices [len(meal_df)-12 : len(meal_df)].
predicted = forecast['yhat'][len(meal_df)-12 : len(meal_df)].values

mae_prophet = mean_absolute_error(actual, predicted)
# Calculate RMSE by taking the square root of the MSE
mse_prophet = mean_squared_error(actual, predicted)
rmse_prophet = np.sqrt(mse_prophet)

print(f"Prophet MAE: {mae_prophet:.2f}")
print(f"Prophet RMSE: {rmse_prophet:.2f}")

# -----------------------------
# ARIMA Model
# -----------------------------
print("\nRunning ARIMA model...")

# Fit the ARIMA model on the entire dataset for comparison
# Note: auto_arima will select the best ARIMA model based on the data
arima_model = auto_arima(meal_df['y'], seasonal=True, m=4, trace=True, error_action='ignore', suppress_warnings=True)
print(arima_model.summary())

# Forecast next 12 weeks
arima_forecast_future = arima_model.predict(n_periods=12)

# Plot ARIMA forecast (actuals + future forecast)
plt.figure(figsize=(10, 5))
plt.plot(meal_df['ds'], meal_df['y'], label='Actual')
# Create date range for the 12 future forecast points
future_dates = pd.date_range(meal_df['ds'].max() + pd.Timedelta(weeks=1), periods=12, freq='W')
plt.plot(future_dates,
         arima_forecast_future, label='ARIMA Forecast', color='red')
plt.title(f"Forecast for Meal ID {meal_id} using ARIMA")
plt.xlabel("Date")
plt.ylabel("Orders")
plt.legend()
plt.grid(True)
plt.show()


# ARIMA Evaluation
# To evaluate ARIMA on the last 12 actual weeks, we need to make predictions
# for those specific time points using the trained model.
# This is done by passing the number of steps to predict, which is the length of the actual data
# or specifically the last 12 steps for evaluation purposes.
# In this case, we can use the model to predict the last 12 data points from the training set.
arima_predicted_actual_weeks = arima_model.predict(n_periods=len(meal_df))[-12:]

# Calculate metrics using the actual values and the ARIMA predictions for those same weeks
mae_arima = mean_absolute_error(actual, arima_predicted_actual_weeks)
# Calculate RMSE for ARIMA by taking the square root of the MSE
mse_arima = mean_squared_error(actual, arima_predicted_actual_weeks)
rmse_arima = np.sqrt(mse_arima)

print(f"ARIMA MAE: {mae_arima:.2f}")
print(f"ARIMA RMSE: {rmse_arima:.2f}")
```
